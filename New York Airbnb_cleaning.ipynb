```python
import pandas as pd
import datetime
```

```python
file=pd.read_csv('AB_NYC_2019.csv')
```

```python
df=pd.DataFrame(file)
```

```python
df.dropna(inplace=True)
```

```python
df['last_review']=pd.to_datetime(df['last_review'])
```

```python
df['year']=df['last_review'].dt.year
```

```python
df
```

```python
df=df[df['year']>=2019]
```

```python
df=df[df['availability_365']>10]
```

```python
df.to_excel('AB_NYC.xlsx')
```
